```python
import yfinance as yf
import pandas as pd
import datetime
```

```python
tic = yf.Ticker('^GSPC') #Ã­ndice S&P500 ^GSPC
print(tic)
tic_data=tic.history(period='max')
print(tic_data.info)
```

```python
tic_data.index=tic_data.index.date
tic_data
```

```python
tic_data.index
```

```python
ind_heb=[]
```

```python
for i in list(range(len(tic_data))):
    d=tic_data.index[i]
    ind_heb.append(dates.HebrewDate.from_pydate(d))
first=ind_heb[:5]
last=ind_heb[-5:]
print('ind_heb first values:',first)
print('ind_heb last values:',last)
```

```python
tic_heb=tic_data.copy()
tic_heb.index=ind_heb
tic_heb
```

```python
start=tic_heb.index[0]
start
```

```python
print(start.year,type(start.year))
print(start.month,type(start.month))
print(start.day,type(start.day))
```

```python
sy=start.year
print(sy,type(sy))
```

```python
tic_year_close=pd.DataFrame(columns=['Value'])
tic_year_close.index.name='Year'
tic_year_close
```

```python
for i in list(range(len(tic_heb))):
    index_date=tic_heb.index[i]
    heb_year=index_date.year
    if heb_year!=sy:
        yprev=tic_heb.index[i-1].year
        value=round(tic_heb['Close'].iloc[i-1],2)
        tic_year_close.loc[yprev]=[value]
        sy=heb_year
tic_year_close
```

```python
tic_year_close['Variation (%)']=list(range(len(tic_year_close)))
tic_year_close['Variation (%)']=tic_year_close['Variation (%)'].astype(float)
#tic_year_close
```

```python
for q in list(range(len(tic_year_close))):
    year_to_access = tic_year_close.index[q]
    if q==0:
        tic_year_close.loc[year_to_access,'Variation (%)']=round((tic_year_close.loc[year_to_access,'Value']/tic_heb.iloc[0]['Close']-1)*100,1)
    else:
        year_previous = tic_year_close.index[q-1]
        tic_year_close.loc[year_to_access,'Variation (%)']=round((tic_year_close.loc[year_to_access,'Value']/tic_year_close.loc[year_previous,'Value']-1)*100,1)
tic_year_close
```

```python
print(sum(tic_year_close['Variation (%)']>0),'positive years')
print(sum(tic_year_close['Variation (%)']==0),'zero returns years')
print(sum(tic_year_close['Variation (%)']<0),'negative years')
```
